## Mathematically Calculate bivariate probability distribution

```python
import numpy as np
X = np.array([[13.04681517, 14.74115241],
              [13.40852019, 13.7632696 ],
              [14.19591481, 15.85318113],
              [14.91470077, 16.17425987]])

def est_gaus_par(X):
    mu = np.mean(X,axis=0)
    sig = np.cov(X,rowvar=0)
    return mu,sig

mu,sigma = est_gaus_par(X)
print("MEAN\n",mu)
print("COVARIANCE\n",sigma)
def est_mult_gaus(X,mu,sigma):
    m = len(mu)
    sigma2=sigma
    #sigma2 = np.diag(sigma)
    X = X-mu.T
    p = 1/((2*np.pi)**(m/2)*np.linalg.det(sigma2)**(0.5))*np.exp(-0.5*np.sum(X.dot(np.linalg.pinv(sigma2))*X,axis=1))
    #p=1
    return p

p = est_mult_gaus(X, mu, sigma)
print("\nP(X)\n",p)
```

```python
import numpy as np
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D

from scipy.stats import multivariate_normal

X=([13.04681517,13.40852019,14.19591481,14.91470077,])
Y=([14.74115241,13.7632696,15.85318113,16.17425987,])


X, Y = np.meshgrid(X,Y)
print("X.shape")
print("\nX.Shape\n",X.shape)
print("\nX.Shape\n",Y.shape)

X_mean = np.mean(X); Y_mean = np.mean(Y)
print("\nX_mean\n",X_mean)
print("\nY_mean\n",Y_mean)

X_var = np.var(X); Y_var = np.var(Y)

print("\nX_var\n",X_var)
print("\nY_var\n",Y_var)

pos = np.empty(X.shape+(2,))
pos[:,:,0]=X

pos[:,:,1]=Y
rv = multivariate_normal([X_mean, Y_mean],[[X_var, 0], [0, Y_var]])
#rv = multivariate_normal(mu,cov)

print(type(rv))
fig = plt.figure()

ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, rv.pdf(pos), cmap="plasma")
#ax.plot_surface(x, y, p, cmap="plasma")

plt.show()

```
