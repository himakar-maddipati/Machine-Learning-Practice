##1. Use the k-means algorithm and Euclidean distance to cluster the following 5 data points into 2 clusters: A1=(2,5), A2=(5,8), A3=(7,5), A4=(1,2), A5=(4,9). Suppose that the initial centroids (centers of each cluster) are A2 and A4. Using k-means,cluster the 5 points and show the followings for one iteration only:

#(a) Show step-by-step the performed calculations to cluster the 5 points.

```python
#import libraries
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 28})
%matplotlib inline
```

```python
#make array from Given data 
X = np.array([[2, 5], [5, 8], [7, 5],[1, 2], [4, 9]])
```

```python
#Initial Centroids
centroids = np.array([[1, 2],[5, 8]])
```

###CALCULATIONS

```python
#list to store cluster1 points
c1_points = []
#list to store cluster2 points
c2_points = []
#list to store  clusters types
clusters_list = []
```

### calculations  to cluster five points(one Iteration)

```python
for p in X:
    x1, y1 = p[0], p[1]
    euclidean_dist_c1 = np.sqrt((centroids[0,0] - x1)**2 + (centroids[0,1] - y1)**2)
    print("point[",x1,",",y1,"] Euclidean distance from Cluster1: ",euclidean_dist_c1)
    euclidean_dist_c2 = np.sqrt((centroids[1,0] - x1)**2 + (centroids[1,1] - y1)**2)
    print("point[",x1,",",y1,"] Euclidean distance from Cluster2: ",euclidean_dist_c2)
    if euclidean_dist_c1 < euclidean_dist_c2:
        print("point[",x1,",",y1,"] is in Cluster1 as its Euclidean distance from Cluster1 <Cluster2 \n")
        c1_points.append(p)
        clusters_list.append('1')

    else:
        print("point[",x1,",",y1,"] is in Cluster2 as its Euclidean distance from Cluster2 <Cluster1 \n")
        c2_points.append(p)
        clusters_list.append('2')
```

```python
arr1=np.array(c1_points)
arr2=np.array(c2_points)
print("\n POINTS IN CLUSTER1 \n")
print(np.array2string(arr1, separator=','))
print("\n POINTS IN CLUSTER2\n ")
print(np.array2string(arr2,separator=","))
print("\n CLUSTERS LIST",clusters_list)
```

```python
#calculate centroid of cluster1(mean of points in cluster1)
c1_center = [np.array(c1_points)[:, 0].mean(), np.array(c1_points)[:, 1].mean()]
#calculate centroid of cluster2(mean of points in cluster2)
c2_center = [np.array(c2_points)[:, 0].mean(), np.array(c2_points)[:, 1].mean()]
```

```python
#print new centroids for both clusters
new_centers=np.array([c1_center, c2_center])
print("new centroids:\n")
print(new_centers)
```

#(c) Calculate the silhouette score and WSS score.

###Silhouette Score Formula

The silhouette Score is a measure of how similar a data point is within-cluster  when compared to other clusters.

Silhouette Score = (a-b)/max(a,b)

where

a=  the average distance between all clusters.

b=  the average distance between each point within a cluster.



```python
from sklearn.metrics import silhouette_score
silhouttescore=[]
for i in range(2,5):
  kmeans = KMeans(n_clusters=i, random_state=24).fit_predict(X)
  silhouttescore.append(silhouette_score(X,kmeans))
```

```python
print(silhouttescore)
```

```python
x = [2,3,4]
plt.figure(figsize = (8,6))
plt.axvline(2, linestyle='--', color='r')
plt.plot(x, silhouttescore)
plt.xlabel('Number of Clusters',fontsize=16)
plt.ylabel('Silhouette',fontsize=16)
plt.xticks([2,3,4])
plt.title("Silhouette score vs No of Clusters",fontsize=18)
plt.show()
```

### Inference from Silhouette score
From the above plot ,Silhouette score  Maximum value is at two clusters.Hence optimum clusters for these data points is TWO

###WSS Score Formula
WSS  is WITHIN cluster SUM of SQUARES. It is sum of the squared differences between all cluster points and centroids.

WCSS=$Σ(P_i{_1}-Centroid_1)^2+...+Σ(P_i{_n}-Centroid_n)^2$

where

$P_i{_1}$ points in $cluster_1$ with $centroid_1$

$P_i{_n}$ points in $cluster_n$ with $centroid_n$



```python
from sklearn.cluster import KMeans
wcss = []  
for i in range(1, 6): 
  kmeans = KMeans(n_clusters = i, init = 'random', random_state = 24)
  kmeans.fit(X) 
  wcss.append(kmeans.inertia_)
```

```python
print("WSS Scores for different clusters\n")
print(wcss)
```

```python
plt.figure(figsize = (8,6))
plt.plot(range(1, 6), wcss) 
plt.axvline(2, linestyle='--', color='r')
plt.xlabel('Number of clusters',fontsize=16) 
plt.ylabel('WSS',fontsize=16) 
plt.title("WSS score vs No of Clusters",fontsize=18)
plt.show()
```

### Inference from WSS score
WSS score plot shows Elbow at two clusters.Hence optimum clusters for these data points is TWO

###4(b)Wrapper Methods (Forward or Backward Feature Elimination, Recursive Feature Elimination etc.).

```python
#function for recursive feature elimination
def RFE_FeatureSelect(model,features):
  rfe = RFE(estimator= model, n_features_to_select=features,step=1)
  rfe.fit(X_train, y_train)
  summary_f = str(rfe.get_feature_names_out()) #RFE selected feature names
  X_train1=rfe.transform(X_train)
  model.fit(X_train1,y_train)
  rfe.fit(X_test,y_test) 
  X_test1=rfe.transform(X_test)
  pred_clf=model.predict(X_test1)
  acc=accuracy_score(y_test,pred_clf)*100
  return summary_f,acc
#returns selected features and accuracy
```

```python
#function for Forward or Backward Feature Elimination
def FeatureSelect(model,dir,features):
  sfs1 = SequentialFeatureSelector(LR_clf, direction=dir, n_features_to_select= features)
  sfs1.fit(X_train, y_train)
  summary_f = str(sfs1.get_feature_names_out()) #Forward or Backward Feature Elimination selected feature names
  X_train1=sfs1.transform(X_train)
  model.fit(X_train1,y_train)
  sfs1.fit(X_test,y_test)
  X_test1=sfs1.transform(X_test)
  pred_clf=model.predict(X_test1)
  acc=accuracy_score(y_test,pred_clf)*100
  return summary_f,acc
#returns selected features and accuracy
```

```python
def plot_features_accuracy(clf1_acc,dir):
  x=[1,2,3,4,5,6,7]
  plt.figure(figsize = (8,6))
  plt.plot(x, clf1_acc, label=dir)
  plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
  plt.xlabel('Number of Features',fontsize=16)
  plt.ylabel('Accuracy',fontsize=16)
  plt.xticks([1,2,3,4,5,6,7])
  plt.title(dir+" Feature Selection vs Accuracy",fontsize=18)
  plt.grid()
  plt.show()
```

```python
from sklearn.cluster import DBSCAN
dbs_collection = dict(epsilon=[],minpoints=[],clust=[],silhouette_score=[])
for epsilon in np.arange(0.3,0.7+0.05,0.05):
    for minpoints in range(2,15+1):
        db = DBSCAN(eps=epsilon,min_samples=minpoints)# DBScan model
        db.fit(X)# fit data
        core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
        core_samples_mask[db.core_sample_indices_] = True
        labels = db.labels_
        clusters = len(set(labels)) - (1 if -1 in labels else 0 ) #-1 indicates noisy data point
        if clusters == 1:
            pass
        else:
            avg_silhouette =silhouette_score(X,labels)
            dbs_collection['epsilon'].append(epsilon)
            dbs_collection['minpoints'].append(minpoints)
            dbs_collection['clust'].append(clusters)
            dbs_collection['silhouette_score'].append(avg_silhouette)
```

### Top Ten list of epsilon,minpoints,clusters and Silhouette score sorted by Silhouette score

```python
dbs_collection_df = pd.DataFrame(dbs_collection)
dbs_collection_df.sort_values(by=['clust','silhouette_score'],ascending=[True,False],inplace=True)
dbs_collection_df = dbs_collection_df.iloc[:10,:]
dbs_collection_df
```
