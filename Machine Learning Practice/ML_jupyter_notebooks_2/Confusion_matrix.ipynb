# CONFUSION MATRIX



```python
import numpy as np
import matplotlib.pyplot as plt

x_index = ['PRED NEG', 'PRED POS', '', '']
y_index=['ACT NEG=60','ACT POS=105','','']

data = np.random.random((4,4))
plt.rc('figure', figsize=(12, 10))

fig = plt.figure()
ax = fig.add_subplot(111)

cax = ax.matshow(data, cmap='Reds')

plt.text(-0.25, 0, 'TN=50', fontsize = 16)
plt.text(0.75, 0, 'FP=10', fontsize = 16)
plt.text(1.55, 0, 'TNR=50/60\n(Specificity)', fontsize = 16)
plt.text(2.75, 0, 'FPR=10/60', fontsize = 16)

plt.arrow(0.0, -0.25, 1.40, 0, width = 0.05)
plt.arrow(0.75, 0.25, 2, 0, width = 0.05)

plt.text(-0.25, 1, 'FN=5', fontsize = 16)
plt.text(0.75, 1, 'TP=100', fontsize = 16)
plt.text(1.55, 1, 'FNR=5/105', fontsize = 16)
plt.text(2.55, 1, 'TPR=100/105\n(Sensitivity\nRecall)', fontsize = 16)
plt.arrow(0.0, 0.75, 1.40, 0, width = 0.05)
plt.arrow(0.75, 1.25, 2, 0, width = 0.05)

plt.text(1.55, 2, '-LL=TNR/FNR', fontsize = 16)
plt.text(2.55, 2, '+LL=TPR/FPR', fontsize = 16)
plt.text(2.55, 3, 'ODDS=+LL/-LL', fontsize = 16)
#plt.arrow(0.0, -0.25, 1.40, 0, width = 0.05)
#plt.arrow(0.75, 0.25, 2, 0, width = 0.05)

xaxis = np.arange(4)
ax.set_xticks(xaxis)
ax.set_yticks(xaxis)
ax.set_xticklabels(x_index,fontweight='bold',fontsize = 16)
ax.set_yticklabels(y_index,fontweight='bold',fontsize = 16)

plt.show()
```

$True\:NEG\:RATE\:(SPECIFICITY)=\frac{TRUE\:NEG}{TRUE\:NEG+FALSE\:POS}$

$\;\;\;\;\;\;$

$FALSE\:NEG\:RATE\:(MISS RATE)=\frac{FALSE\:NEG}{FALSE\:NEG+TRUE\:POS}$

$\;\;\;\;\;\;$


$False\:POS\:RATE\:(FALL OUT)=\frac{FALSE\:POS}{TRUE\:NEG+FALSE\:POS}$

$\;\;\;\;\;\;$

$True\:POS\:RATE\:(SENSITIVITY,RECALL)=\frac{TRUE\:POS}{FALSE\:NEG+TRUE\:POS}$

$\;\;\;\;\;\;$



```python
x_index = ['PRED NEG', 'PRED POS', '', '']
y_index=['ACT NEG=60','ACT POS=105','','']

data = np.random.random((4,4))
plt.rc('figure', figsize=(12, 10))

fig = plt.figure()
ax = fig.add_subplot(111)

cax = ax.matshow(data, cmap='Reds')

plt.text(-0.25, 0, 'TN=50', fontsize = 16)
plt.text(0.75, 0, 'FP=10', fontsize = 16)
plt.text(1.55, 0, 'TNR=50/60\n(Specificity)', fontsize = 16)
plt.text(2.75, 0, 'FPR=10/60', fontsize = 16)


plt.text(-0.25, 1, 'FN=5', fontsize = 16)
plt.text(0.75, 1, 'TP=100', fontsize = 16)
plt.text(1.55, 1, 'FNR=5/105', fontsize = 16)
plt.text(2.55, 1, 'TPR=100/105\n(Sensitivity\nRecall)', fontsize = 16)

plt.text(1.55, 2, '-LL=TNR/FNR', fontsize = 16)
plt.text(2.55, 2, '+LL=TPR/FPR', fontsize = 16)
plt.text(2.55, 3, 'ODDS=+LL/-LL', fontsize = 16)
plt.arrow(2.25, 0, 0, 1.5, width = 0.05)
plt.arrow(3.25, 1, 0, 0.5, width = 0.05)
plt.arrow(3.25, 2.10, 0, 0.5, width = 0.05)

xaxis = np.arange(4)
ax.set_xticks(xaxis)
ax.set_yticks(xaxis)
ax.set_xticklabels(x_index,fontweight='bold',fontsize = 16)
ax.set_yticklabels(y_index,fontweight='bold',fontsize = 16)

plt.show()
```


$-\:LOG\:LIKELIHOOD=\frac{TRUE\:NEG\:RATE}{FALSE\:NEG\:RATE}$

$\;\;\;\;\;\;$



$+\:LOG\:LIKELIHOOD=\frac{TRUE\:POS\:RATE}{FALSE\:POS\:RATE}$

$\;\;\;\;\;\;$

$DIAGNOSTIC\:ODDS=\frac{+\:LOG\:LIKELIHOOD}{-\:LOG\:LIKELIHOOD}$



```python
x_index = ['PRED NEG', 'PRED POS', '', '']
y_index=['ACT NEG=60','ACT POS=105',' ','']

data = np.random.random((4,4))
plt.rc('figure', figsize=(12, 10))

fig = plt.figure()
ax = fig.add_subplot(111)

cax = ax.matshow(data, cmap='Reds')

#fig.colorbar(cax)
plt.text(-0.25, 0, 'TN=50', fontsize = 16)
plt.text(-0.25, 1, 'FN=5', fontsize = 16)
plt.text(-0.25, 2, 'NPR=50/55', fontsize = 16)
plt.text(-0.25, 3, 'FOR=5/55', fontsize = 16)
plt.arrow(-0.35, 0.25, 0.0, 1.5, width = 0.05)
plt.arrow(0.45, 1.25, 0, 1.5, width = 0.05)

plt.text(0.75, 0, 'FP=10', fontsize = 16)
plt.text(0.75, 1, 'TP=100', fontsize = 16)
plt.text(0.75, 2, 'FDR=10/110', fontsize = 16)
plt.text(0.75, 3, 'PPR=100/110\n(Precision)', fontsize = 16)
plt.arrow(0.65, 0.25, 0.0, 1.5, width = 0.05)
plt.arrow(1.45, 1.25, 0, 1.5, width = 0.05)

xaxis = np.arange(4)
ax.set_xticks(xaxis)
ax.set_yticks(xaxis)
ax.set_xticklabels(x_index,fontweight='bold',fontsize = 16)
ax.set_yticklabels(y_index,fontweight='bold',fontsize = 16)

plt.show()
```

$NEG\:PREDICTIVE\:VALUE\:=\frac{TRUE\:NEG}{TRUE\:NEG+FALSE\:POS}$

$\;\;\;\;\;\;$

$FALSE\:OMMISSION\:RATE\:=\frac{FALSE\:NEG}{FALSE\:NEG+TRUE\:POS}$

$\;\;\;\;\;\;$


$False\:DISCOVERY\:RATE\:=\frac{FALSE\:POS}{TRUE\:NEG+FALSE\:POS}$

$\;\;\;\;\;\;$

$POS\:PREDICTIVE\:RATE\:(PRECISION)=\frac{TRUE\:POS}{FALSE\:NEG+TRUE\:POS}$

$\;\;\;\;\;\;$






```python
x_index = ['PRED NEG', 'PRED POS', '', '']
y_index=['ACT NEG=60','ACT POS=105','NULL ERROR=60/105','']

data = np.random.random((4,4))
plt.rc('figure', figsize=(12, 10))

fig = plt.figure()
ax = fig.add_subplot(111)

cax = ax.matshow(data, cmap='Reds')

plt.text(-0.25, 0, 'TN=50', fontsize = 16)
plt.text(-0.25, 1, 'FN=5', fontsize = 16)
plt.text(1.55, 0, 'MISS CLASSIFICATION=15/165', fontsize = 16,color='green')

plt.text(0.75, 0, 'FP=10', fontsize = 16)
plt.text(0.75, 1, 'TP=100', fontsize = 16)
plt.text(1.55, 1, 'PREVALANCE=105/165', fontsize = 16)
plt.text(1.55, 2, 'ACCURACY=150/165', fontsize = 16,color='blue')
plt.arrow(0.0, 0.75, 1.40, 0, width = 0.05,color='black')
plt.arrow(0, 0, 2, 1.75, width = 0.05,color='blue')
plt.arrow(0.2, 1, 1.2, -1, width = 0.05,color='green')



xaxis = np.arange(4)
print(xaxis)
ax.set_xticks(xaxis)
ax.set_yticks(xaxis)
ax.set_xticklabels(x_index,fontweight='bold',fontsize = 16)
ax.set_yticklabels(y_index,fontweight='bold',fontsize = 16)

plt.show()
```


$\:ACCURACY=\frac{TRUE\:POS\:+\:TRUE\:NEG}{TRUE\:POS\:+FALSE\:POS\:+\:TRUE\:NEG+FALSE\:NEG\:}$

$\;\;\;\;\;\;$

$PREVALENCE=\frac{TRUE\:POS\:+\:FALSE\:NEG}{TRUE\:POS\:+FALSE\:POS\:+\:TRUE\:NEG+FALSE\:NEG\:}$

$\;\;\;\;\;\;$

$MISS\:CLASSIFICATION\:RATE=\frac{FALSE\:NEG\:+\:FALSE\:POS}{TRUE\:POS\:+FALSE\:POS\:+\:TRUE\:NEG+FALSE\:NEG\:}$
